# Overview
Repeat some of the previous steps to create a KG from an independent source.
This is for the purposes of triangulation / sense-checking /bringing in new stakeholders.

The KG can come from :
- whiteboard session with other stakeholders
- a project framework
- key strategy document.

For this example, we have extracted keywords from a paper on success characteristics for one particular type of nuclear projects: Decommissioning.

The paper is 'Diletta Colette Invernizzi, Giorgio Locatelli & Naomi J. Brookes (2020): Characterising nuclear decommissioning projects: an investigation of the project characteristics that affect the project performance, Construction Management and Economics'
